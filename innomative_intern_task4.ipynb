# Exploratory Data Analysis

```python
import pandas as pd

data = pd.read_csv('task4_data.csv')

print(data.head())

print(f"Dataset contains {data.shape[0]} rows and {data.shape[1]} columns")

print(data.describe())

```

```python
print(data.isnull().sum())
```

## Univariate Analysis

a. Distribution of Salary

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.histplot(data['Salary'], kde=True)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

b. Boxplot to Identify Outliers in Salary

```python
plt.figure(figsize=(10,6))
sns.boxplot(data['Salary'])
plt.title('Boxplot of Salary')
plt.show()
```

c. Count Plot for Categorical Variables (e.g., Gender, Specialization)


```python
# Count plot for Gender
plt.figure(figsize=(10,6))
sns.countplot(data['Gender'])
plt.title('Gender Distribution')
plt.show()

# Count plot for Specialization
plt.figure(figsize=(12,6))
sns.countplot(y='Specialization', data=data)
plt.title('Specialization Distribution')
plt.show()
```

d. Outliers in Other Numerical Columns


```python
plt.figure(figsize=(10,6))
sns.boxplot(data['10percentage'])
plt.title('10th Percentage Boxplot')
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(data['collegeGPA'])
plt.title('College GPA Boxplot')
plt.show()
```

e. Frequency Distribution of Other Numerical Columns


```python
# Histogram for 10percentage
plt.figure(figsize=(10,6))
sns.histplot(data['10percentage'], kde=True)
plt.title('10th Percentage Distribution')
plt.show()

# Histogram for college GPA
plt.figure(figsize=(10,6))
sns.histplot(data['collegeGPA'], kde=True)
plt.title('College GPA Distribution')
plt.show()
```

## Bivariate Analysis

a. Scatter Plot Between Salary and Numerical Variables 

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='10percentage', y='Salary', data=data)
plt.title('10th Percentage vs Salary')
plt.show()

plt.figure(figsize=(10,6))
sns.scatterplot(x='collegeGPA', y='Salary', data=data)
plt.title('College GPA vs Salary')
plt.show()
```

b. Boxplots for Salary vs Categorical Variables 

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Gender', y='Salary', data=data)
plt.title('Salary by Gender')
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(x='Specialization', y='Salary', data=data)
plt.xticks(rotation=90)
plt.title('Salary by Specialization')
plt.show()
```

c. Stacked Bar Plot for Categorical Variables

```python
pd.crosstab(data['Gender'], data['Specialization']).plot(kind='bar', stacked=True, figsize=(12,6))
plt.title('Gender vs Specialization')
plt.show()
```

## Research Questions

a. Salary Range Test for Fresh Graduates


```python
job_roles = ['Programming Analyst', 'Software Engineer', 'Hardware Engineer', 'Associate Engineer']
fresh_grads = data[data['Designation'].isin(job_roles)]
```

```python
salary_mean = fresh_grads['Salary'].mean()
print(f"Average Salary for fresh graduates in the selected roles: {salary_mean}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.boxplot(x='Designation', y='Salary', data=fresh_grads)
plt.xticks(rotation=90)  
plt.title('Salary for Selected Job Roles')
plt.show()
```

b. Relationship Between Gender and Specialization


```python
from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(data['Gender'], data['Specialization'])

chi2, p, dof, ex = chi2_contingency(contingency_table)
print(f"Chi-Square Test result: Chi2 = {chi2}, p-value = {p}")

if p < 0.05:
    print("There is a significant relationship between Gender and Specialization")
else:
    print("There is no significant relationship between Gender and Specialization")
```
